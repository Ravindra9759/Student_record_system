```python
import json 
import csv
import os
from datetime import datetime
from tabulate import tabulate
data_json = "student.json"
data_csv = "Student.csv"

# the load file in JSON and CSV.

def load_data(file_t = "json"):
    if file_t == "json":
        if os.path.exists(data_json):
            with open(data_json,"r") as f:
                return json.load(f)
        return []
    else:
        if os.path.exists(data_csv):
            with open(data_csv,"r") as f:
                      reader = csv.DictReader(f)
                      return list(reader)
        return []
# Save the data in JSON/CSV Function.
def save_data(data,file_type="json"):
    if file_type == "json":
        with open(data_json, "w") as f:
            json.dump(data,f,indent=9)
    else:
        with open(data_csv,"w",newline= "") as f:
            writer = csv.DictWriter(f, fieldnames = ["roll","name","F_name","College_name","Course","Started_date","End_date","Address","Mobile_Number"])
            writer.writeheader()
            writer.writerows(data)

# Create a date time function
def validate_date(date_text):
    try:
        datetime.strptime(date_text,"%y-%m-%d")
        return True
    except ValueError:
        return False
        
# Validate Function in Student Records.
def validate_student(roll,name,F_name,college_name,Course,Started_date,End_date,Address,Mobile_Number,students):
    if not roll.isdigit():
        print("âŒ Roll Number must be numeric!")
        
        return False
    if any(s["roll"] == roll for s in students):
        print("âŒ roll number already exists!")
        return False
    if not name.isalpha():
        print("âŒ Name must contain only alphabets!")
        return False
    if not F_name.isalpha():
        print("âŒ Father Name must be numeric!")
        return False
    if not college_name.isalpha():
        print("âŒ College must contain only alphabets!")
        return False
    if not Course.isalpha():
        print("âŒ Course must contain only alphabets!")
        return False
    if not validate_date(Started_date):
        print("âŒ Started must contain only date!")
        return False
    if not validate_date(End_date):
        print("âŒ End_date must contain only date!")
        return False
    if not Address.replace(" ","").isalpha():
        print("âŒ Address must contain only alphabets!")
        return False
    if not Mobile_Number.isdigit():
        print("âŒ Mobile_Number must contain only numeric!")
        return False
    return True

def add_student(file_type):
    students = load_data(file_type)
    roll = input("Enter Roll Number : ")
    name = input("Enter Name : ")
    F_name = input("Enter Father Name : ")
    College_name = input("Enter College_name : ")
    Course = input("Enter Course : ")
    Started_date = input("Enter Started_date : ")
    End_date = input("Enter End_date : ")
    Address = input("Enter Address : ")
    Mobile_Number = input("Enter Mobile_Number : ")
    if validate_student(roll,name, F_name,College_name,Course,Started_date,End_date,Address,Mobile_Number,students):
        students.append({"roll" : roll, "name" : name, "F_name" : F_name, "College_name" : College_name, "Course" : Course, "Started_date" : Started_date, "End_date" : End_date, "Address" : Address, "Mobile_Number" : Mobile_Number})
        save_data(students,file_type)
        print("âœ… Student added Successfully!")

def view_student(file_type):
    students = load_data(file_type)
    if students:
        print(tabulate(students, headers="keys", tablefmt = "grid"))
    else:
        print("Not record fund!")
def update_student(file_type):
    students = load_data(file_type)
    roll = input("Enter Roll Number to Update : ")
    for s in students:
        if s["roll"] == roll:
            s["name"] = input("Enter the Name :")
            s["F_name"] = input("Enter the Father Name :")
            s["College_name"] = input("Enter the College Name :")
            s["Course"] = input("Enter the Course Name :")
            s["Started_date"] = input("Enter the Started Date(y-m-d) :")
            s["End_date"] = input("Enter the End Date (y-m-d) :")
            s["Address"] = input("Enter the Address :")
            s["Mobile_Number"] = input("Enter the Mobile Number :")
            save_data(students , file_type)
            print("âœ… Record Update")
            return
    print("âŒ Student Not Found!")

def delete_student(file_type):
    students = load_data(file_type)
    roll = input("Enter Roll Number to Delete : ")
    new_students = [s for s in students if s["roll"] != roll]
    if len(new_students) != len(students):
        save_data(new_students, file_type)
        print("âœ… Record Delete")
    else:
        print("âŒ Student  Not Found!")

def search_student(file_type):
    students = load_data(file_type)
    keyword = input("Enter the name or roll to Search !")
    results = [s for s in students if keyword.lower() in s ["name"].lower()  or s["roll"] == keyword]
    if results:
        print(tabulate(results, headers="keys", tablefmt="grid"))
    else:
        print("No matching record found !")

def menu():
    print("\nğŸ§‘â€ğŸ“Student Record Management System ")
    print("1. Add Student")
    print("2. View Student")
    print("3. Update Student")
    print("4. Delete Student")
    print("5. Search Student")
    print("6. Exit")

def main():
    file_type = input("Choose Storage type (json/csv) : ").strip().lower()
    while True:
        menu()
        choice = input("Enter choice : ")
        if choice == "1":
            add_student(file_type)
        elif choice == "2":
            view_student(file_type)
        elif choice == "3":
            update_student(file_type)
        elif choice == "4":
            delete_student(file_type)
            print("âŒ Invail Choice !")

        elif choice == "5":
            search_student(file_type)
        elif choice == "6":
            print("ğŸ‘‹ Exiting Program... !")
        else:
            print("ğŸ˜’ Invalid Choice!")
if __name__ == "__main__":
    main()         
```
